```python
import pandas as pd

car_row = pd.read_csv('data/car.data', header=None)
car_row.head()
```

```python
car_row.info()
```

```python
#컬럼 이름을 추가하기
column_name = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'class']
data = car_row.values.tolist()
car_df = pd.DataFrame(data, columns=column_name)

car_df.head()
```

```python
car_df.info()
```

```python
#문자 데이터를 숫자로 바꾸는 2가지 방안에 대해서 설명
#모든 문자 카테고리 데이터를 숫자(one-hot encoding)으로 변경
car_df = pd.get_dummies(car_df, dtype='int') # columns=None 파라미터 생략
car_df.head()
```

```python
car_df.info()
```

```python
# feature와 target 으로 분리
X = car_df.iloc[:, :21]
y = car_df.iloc[:, 21:]
```

```python
X.info()
```

```python
y.info()
```

```python
# 훈련데이터 테스트 데이터 분리
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True)
```

```python
# 모델
from tensorflow.keras import models, layers

model = models.Sequential()
model.add(layers.Dense(42, input_dim=21, activation='relu'))
model.add(layers.Dense(21, activation='relu'))
model.add(layers.Dense(12, activation='relu'))
model.add(layers.Dense(4, activation='softmax'))
model.summary()
```

```python
#학습
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=20, batch_size=10)
```

```python
# 학습과정의 정확도 시각화
import numpy as np
import matplotlib.pyplot as plt

y_vacc = history.history['val_accuracy']

y_acc = history.history['accuracy']


x_len = np.arange(len(y_acc))
plt.plot(x_len, y_vacc, marker='.', c="red", label='Testset_accuracy')
plt.plot(x_len, y_acc, marker='.', c="blue", label='Trainset_accuracy')

plt.legend()
plt.grid()
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.show()
```

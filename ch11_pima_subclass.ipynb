```python
import pandas as pd
```

```python
df = pd.read_csv('data/pima-indians-diabetes3.csv')
```

```python
dataset = df.values
X = dataset[:, :8]
y = dataset[:, 8]
```

```python
model_name = 'pima_indian_diabetes'
BATCH_SIZE = 10
```

```python
params = {'model_name': model_name, 
          'input_dimension': 8
          ,'hidden_dimension_01': 12
          ,'hidden_dimension_02': 8
          ,'output_dimension': 1}
```

```python
from tensorflow.keras import layers
import tensorflow as tf

class PimaIndian(tf.keras.Model):
    def __init__(self, **kargs):
        super(PimaIndian, self).__init__(name=kargs['model_name'])
        #self.input = layers.Input(shape=(kargs['input_dimension']))
        self.fc1 = layers.Dense(kargs['hidden_dimension_01'],
                                input_dim=kargs['input_dimension'],
                                activation='relu')
        self.fc2 = layers.Dense(kargs['hidden_dimension_02'], activation='relu')
        self.fc3 = layers.Dense(kargs['output_dimension'], activation='sigmoid')

    def call(self, x):
        #x = self.input(x)
        x = self.fc1(x)
        x = self.fc2(x)
        x = self.fc3(x)

        return x
```

```python
model = PimaIndian(**params)
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X, y, epochs=100, batch_size=5)
```

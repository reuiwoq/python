```python
years = range(1880, 2011)

pieces = []
columns= ['name','sex','births']

for year in years:
    path = 'data/yob%d.txt' % year
    frame = pd.read_csv(path, names=columns)
    frame['year'] = year
    pieces.append(frame)

names = pd.concat(pieces, ignore_index=True)
```

```python
names.info()
```

```python
names
```

```python
total_births = names.pivot_table('births', index='year',columns='sex', aggfunc='sum')
total_births.tail()
```

```python
import matplotlib.pyplot as plt
total_births.plot(title="Total births by sex and year")
```

```python
def add_prop(group):
    group['prop'] = group.births / group.births.sum()
    return group
    
names = names.groupby(['year','sex'], as_index=False)[['name','sex','births','year']].apply(add_prop)
```

```python
names
```

```python
names.groupby(['year','sex']).prop.sum()
```

```python
def get_top1000(group):
    return group.sort_values(by='births', ascending=False)[:1000]

grouped = names.groupby(['year','sex'])[['name','sex','births','year','prop']]
top1000 = grouped.apply(get_top1000)
```

```python
top1000
```

```python
top1000.reset_index(inplace=True, drop=True)
top1000
```

```python
boys = top1000[top1000.sex == 'M']
girls = top1000[top1000.sex == 'F']
```

```python
total_births = top1000.pivot_table('births', index='year', columns='name', aggfunc='sum')
```

```python
df = boys[boys.year == 1900]
in1900 = df.sort_values(by='prop', ascending=False).prop.cumsum()
in1900.values.searchsorted(0.5)
```

```python
def get_quantile_count(group, q=0.5):
    group = group.sort_values(by='prop', ascending=False)
    return group.prop.cumsum().values.searchsorted(q) + 1 #index가 0부터 시작하기 때문에 개수를 뽑아내기 위해서

diversity = top1000.groupby(['year', 'sex'])[['year', 'sex', 'prop']].apply(get_quantile_count)
```

```python
diversity = diversity.unstack('sex')
diversity
```

```python
get_last_letter = lambda x: x[-1]
last_letters = names.name.map(get_last_letter)
last_letters.name = 'last_letter'
last_letters
```

```python
table = names.pivot_table('births', index=last_letters, columns=['sex', 'year'], aggfunc='sum')
table
```

```python
subtable = table.reindex(columns=[1910, 1960, 2010], level='year')
subtable.head()
```

```python
subtable.sum()
```

```python
letter_prop = subtable / subtable.sum()
letter_prop
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 1, figsize=(10, 8))
plt.subplots_adjust(hspace=0.25)
letter_prop['M'].plot(kind='bar', rot=0, ax=axes[0], title='Male')
letter_prop['F'].plot(kind='bar', rot=0, ax=axes[1], title='Female', legend=False)
```

```python
all_names = pd.Series(top1000.name.unique())
lesley_like = all_names[all_names.str.lower().str.contains('lesl')]
lesley_like
```

```python
filtered = top1000[top1000.name.isin(lesley_like)]
filtered.groupby('name').births.sum()
```

```python
table = filtered.pivot_table('births', index='year', columns='sex', aggfunc='sum')
table
```

```python
table = table.div(table.sum(1), axis=0) # axis=0 means match Series index on
table.head()
```

```python
table.plot(style={'M': 'b-', 'F': 'r--'}, figsize=(10, 8))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# x1, x2 값
X1 = np.arange(-1.0, 1.0, 0.2)  # 원소 수는 10개 
X2 = np.arange(-1.0, 1.0, 0.2)

# 출력값을 저장할 10x10 그리드
Z = np.zeros((10,10))

# x, y 값의 입력 가중치
W = np.array([2.5,3.0])
#W = np.array([1.0,1.0])

# 편향
bias = np.array([0.1])
```

```python
# 그리드맵의 각 그리드별 뉴런의 연산
for i in range(10):
    for j in range(10):
        
        # 입력과 가중치 곱의 합 + 편향
        X = np.array( [ X1[i], X2[j] ] )
        u = np.dot(X, W.T) + bias 
        # 그리드맵에 출력 값 저장
        y = 1/(1+np.exp(-u)) # 시그모이드 함수
        Z[j][i] = y[0]
```

```python
plt.imshow(Z, "gray", vmin=0.0, vmax=1.0)
plt.colorbar()
plt.show()
```
